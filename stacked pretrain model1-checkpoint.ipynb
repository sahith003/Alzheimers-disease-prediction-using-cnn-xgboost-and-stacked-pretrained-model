```python
import warnings 
warnings.filterwarnings('ignore')
from zipfile import ZipFile
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,MaxPooling2D,Conv2D,Flatten,Dropout,GlobalAveragePooling2D,BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras import regularizers,optimizers
import os
from imblearn.over_sampling import SMOTE
```

```python
WORK_DIR = './dataset/'
IMG_SIZE = 224
DIM = (IMG_SIZE, IMG_SIZE)
ZOOM = [.99, 1.01]
BRIGHT_RANGE = [0.8, 1.2]
```

```python
work_dr = ImageDataGenerator(rescale = 1./255, brightness_range=BRIGHT_RANGE, zoom_range=ZOOM, horizontal_flip=True,vertical_flip=True)
data_gen = work_dr.flow_from_directory(directory=WORK_DIR, target_size=DIM, batch_size=2300,shuffle=False)
```

```python
data, data_labels = data_gen.next()
print(data.shape,data_labels.shape)
```

```python
sm = SMOTE(random_state=42)

data1, data_labels1 = sm.fit_resample(data.reshape(-1, IMG_SIZE * IMG_SIZE * 3), data_labels)

data1 = data1.reshape(-1, IMG_SIZE, IMG_SIZE, 3)

print(data1.shape, data_labels1.shape)
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(data1,data_labels1, test_size = 0.2, random_state=42)
train_data, val_data, train_labels, val_labels = train_test_split(train_data, train_labels, test_size = 0.2, random_state=42)
```

```python
from keras.applications.densenet import DenseNet121, preprocess_input as densenet_preprocess
from keras.applications.vgg16 import VGG16,preprocess_input as vgg_preprocess
from tensorflow.keras.applications import ResNet50V2
from tensorflow.keras.applications.resnet_v2 import preprocess_input as resnet_preprocess
from keras.layers.merge import concatenate
from tensorflow.keras.layers import  Lambda
from keras.models import Model
```

```python
input_shape=(224,224,3)
input_layer=tf.keras.Input(shape=input_shape)
```

```python
preprocessor_vgg16=Lambda(vgg_preprocess)(input_layer)
vgg16=VGG16(include_top=False,weights='imagenet',input_shape=input_shape)(preprocessor_vgg16)
preprocessor_densenet=Lambda(densenet_preprocess)(input_layer)
densenet21=DenseNet121(include_top=False,weights='imagenet',input_shape=input_shape)(preprocessor_densenet)
preprocessor_resnet=Lambda(resnet_preprocess)(input_layer)
resnet50v2=ResNet50V2(include_top=False,weights='imagenet',input_shape=input_shape)(preprocessor_resnet)
```

```python
concat=concatenate([vgg16,densenet21,resnet50v2])
model=Model(inputs=input_layer,outputs=concat)
```

```python
x_train=model.predict(train_data)
```

```python
x_val=model.predict(val_data)
```

```python
x_test=model.predict(test_data)
```

```python
def roundoff(arr):
    arr[np.argwhere(arr != arr.max())] = 0
    arr[np.argwhere(arr == arr.max())] = 1
    return arr
```
